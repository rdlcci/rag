```python
import itertools
import pandas as pd
import math


# Read Excel file and get columns as dictionary from sheet 1
input_file = "rag experiments.xlsx"
input_sheet = "Text Experiements"
output_file = "text_experiments.csv"
```

```python
df = pd.read_excel(input_file, sheet_name=input_sheet)
df.head()
```

```python
columns = {col: df[col].dropna().unique().tolist() for col in df.columns[:-6]}
columns
```

```python
# Generate all possible combinations of values across keys in 'columns'
keys = list(columns.keys())
values = [columns[k] if columns[k] else [None] for k in keys]
values
```

```python
math.prod([len(i) for i in values])
```

```python
combinations = list(itertools.product(*values))

# Convert combinations to list of dicts
comb_dicts = [dict(zip(keys, combo)) for combo in combinations]
comb_dicts
```

```python
output = pd.DataFrame(comb_dicts, columns=keys)
output.to_csv(output_file)
```
